```python
import jieba
import pandas as pd
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/kshao/Desktop/weibo-crawler-master/8_22_七夕_NfUYi2Vy8.csv')
res = []
for val in data["comment_content"].values:
    tmp = BeautifulSoup(val,'html.parser').get_text()
    if tmp != '':
        res.append(tmp)
```

```python
res_text = ''
for val in res:
    words = jieba.lcut(val)
    newtxt = ' '.join(words)
    res_text = res_text + ' ' + newtxt
```

```python
with open('./cn_stopwords.txt', 'r') as file:
    lines = [line.strip() for line in file.readlines()]
print(lines)
```

```python
wc = WordCloud(font_path='./msyh.ttf',background_color='white', max_words=500,stopwords=lines).generate(res_text)
```

```python
wc.to_file('8_22_七夕.png')
plt.imshow(wc)
plt.axis('off')
plt.show()
```
